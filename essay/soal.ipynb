```python
#1. 
import math

nama = str(input("Masukkan nama: "))
nim = int(input("Masukkan NIM: "))
usia = int(input("Masukkan usia: "))
angka_favorit = float(input("Masukkan angka favorit: "))

jari_jari = float(angka_favorit)

luaslingkaran = math.pi * jari_jari**2

print(f"\nHalo, {nama}!")
print(f"nim anda: {nim}")
print(f"usia anda: {usia}tahun")
print(f"angka favorit: {angka_favorit}")
print(f"Luas lingkaran dengan jari-jari {angka_favorit} adalah {luaslingkaran:.2f}")
```

```python
# 2. data surabaya
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('data genangan air Surabaya (1).csv')
print(data)
```

```python
# a. a nilai maksimum, minimum dan rata-rata
maksluas = data['Luas genangan'].max()
minluas = data['Luas genangan'].min()
rataluas = data['Luas genangan'].mean()

makstinggi = data['Tinggi genangan'].max()
mintinggi = data['Tinggi genangan'].min()
ratatinggi = data['Tinggi genangan'].mean()

makslama = data['Lama genangan'].max()
minlama = data['Lama genangan'].min()
ratalama = data['Lama genangan'].mean()

print("nilai maksimum, minimum, dan rata-rata:")
print(f"luas genangan : maks: {maksluas}, min: {minluas}, rata-rata: {rataluas}")
print(f"tinggi genangan : maks: {makstinggi}, min: {mintinggi}, rata-rata: {ratatinggi}")
print(f"lama genangan : maks: {makslama}, min: {minlama}, rata-rata: {ratalama}")
```

```python
# b. pie chart
rayon_pie = data['Rayon'].value_counts()
plt.figure(figsize=(8, 6))
plt.pie(rayon_pie, labels=rayon_pie.index, autopct='%1.1f%%', startangle=140)
plt.title('distribusi genangan Rayon')
plt.show()
```

```python
#c.
plt.figure(figsize=(10, 5))

# luas genangan dan skor genangan
plt.scatter(data['Luas genangan'], data['Tinggi genangan'])
plt.title('hubungan luas genangan dan skor genangan')
plt.xlabel('luas Genangan')
plt.ylabel('skor Genangan')
plt.show()

# luas genangan dan lama genangan
plt.scatter(data['Luas genangan'], data['Lama genangan'])
plt.title('hubungan luas genangan dan lama genangan')
plt.xlabel('luas Genangan')
plt.ylabel('lama Genangan')
plt.show()
```

```python
#3. 
import pandas as pd

# a. duplikasi
if data.duplicated().any():
    print("ada data duplikat.")
    data = data.drop_duplicates()  
    print("data duplikat dihapus.")
else:
    print("tidak ada duplikat.")
```

```python
#b. missing value
if data.isnull().any().any():
    print("ada missing value.")
    data = data.fillna(method='ffill')  
    print("missing value sudah diisi nilai sebelumnya.")
else:
    print("tidak ada missing value.")
```

```python
#c. ubah tipe data
data['Kecamatan'] = data['Kecamatan'].astype('category').cat.codes
data['Kelurahan'] = data['Kelurahan'].astype('category').cat.codes
data['Rayon'] = data['Rayon'].astype('category').cat.codes

print("\ndata setelah praproses:")
print(data.head())
```
